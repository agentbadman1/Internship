```python
from bs4 import BeautifulSoup
import requests
```

```python
url = requests.get("https://www.imdb.com/india/top-rated-indian-movies/")
```

```python
url
```

```python
soup = BeautifulSoup(url.content)
```

```python
titles = []

for i in soup.find_all("td",class_="titleColumn"):
    titles.append(i.text.split("\n")[2])
    
titles
```

```python
year = []

for i in soup.find_all("td",class_="titleColumn"):
    year.append(i.text.split("\n")[3])
    
year
```

```python
rating = []

for i in soup.find_all("td",class_="ratingColumn imdbRating"):
    rating.append(i.text.split("\n")[1])
    
rating
```

```python
print(len(titles),len(year),len(rating))
```

```python
titles = titles[:100]
```

```python
len(titles)
```

```python
year = year[:100]
```

```python
rating = rating[:100]
```

```python
print(len(titles),len(year),len(rating))
```

```python
import pandas as pd
```

```python
data = pd.DataFrame({"Name":titles,"Rating":rating,"Year of Release":year})
```
