```python
from bs4 import BeautifulSoup
import requests
```

```python
url = requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc")
```

```python
url
```

```python
soup = BeautifulSoup(url.content)
```

```python
titles = []

for i in soup.find_all("h3",class_="lister-item-header"):
    titles.append(i.text.split("\n")[2])
    
titles
```

```python
year = []

for i in soup.find_all("span",class_="lister-item-year text-muted unbold"):
    year.append(i.text)
    
year
```

```python
rating = []

for i in soup.find_all("div",class_="inline-block ratings-imdb-rating"):
    rating.append(i.text.split("\n")[2])
    
rating
```

```python
url = requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt")
```

```python
url
```

```python
soup = BeautifulSoup(url.content)
```

```python
for i in soup.find_all("h3",class_="lister-item-header"):
    titles.append(i.text.split("\n")[2])
    
titles
```

```python
for i in soup.find_all("span",class_="lister-item-year text-muted unbold"):
    year.append(i.text)
    
year
```

```python
for i in soup.find_all("div",class_="inline-block ratings-imdb-rating"):
    rating.append(i.text.split("\n")[2])
    
rating
```

```python
print(len(titles),len(year),len(rating))
```
