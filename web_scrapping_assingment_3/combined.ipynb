```python

import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings("ignore")
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import NoSuchElementException
driver = webdriver.Chrome(r"chromedriver.exe")
driver.get("https://www.youtube.com/watch?v=kJQP7kiw5Fk")
driver.maximize_window()
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
comment = []
upvote = []
time = []
try:
    comment_tag = driver.find_elements(By.XPATH,'//div[@class="style-scope ytd-expander"]')
    for i in comment_tag[2:502]:
        j = i.text
        comment.append(j)
except NoSuchElementException as e:
    comment.append("-")
len(comment)
500
try:
    upvote_tag = driver.find_elements(By.XPATH,'//span[@class="style-scope ytd-comment-action-buttons-renderer"]')
    for i in upvote_tag[1:1101]:
        j = i.text
        if len(j) != 0:
            upvote.append(j)
except NoSuchElementException as e:
    upvote.append("-")
try:
    time_tag = driver.find_elements(By.XPATH,'//a[@class="yt-simple-endpoint style-scope yt-formatted-string"]')
    for i in time_tag[50:550]:
        j = i.text
        time.append(j)
        
except NoSuchElementException as e:
    time.append("-")
data = pd.DataFrame({"COMMENT":comment,"UPVOTE":upvote,
                    "TIME":time})
data

upvote = upvote[:500]
data.to_csv("youtube top 500 comments.csv")
 
```

```python

import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings("ignore")
from selenium.webdriver.common.by import By
import time
driver = webdriver.Chrome(r"chromedriver.exe")
driver.get("https://www.hostelworld.com/")
driver.maximize_window()
search = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div/div/div[4]/div/div[2]/div/div[1]/div/div/div/input')
search.send_keys("London")
go = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div/div/div[4]/div/div[2]/div/div[1]/div/div/ul/li[2]/div')
go.click()
gogo = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div/div/div[4]/div/div[2]/div/div[5]/button')
gogo.click()
name = []
dist = []
rating = []
total_review = []
dp = []
pp = []
fac = []
propd = []
over = []
price = []
urls = []
name_tag = driver.find_elements(By.XPATH,'//h2[@class="title title-6"]')
for i in name_tag:
    j = i.text
    name.append(j)
dist_tag = driver.find_elements(By.XPATH,'//span[@class="description"]')
for i in dist_tag:
    j = i.text.split(" - ")[1]
    dist.append(j)
rating_tag = driver.find_elements(By.XPATH,'//div[@class="rating rating-summary-container big"]')
for i in rating_tag:
    j = i.text.split("\n")[0]
    rating.append(j)
total_review_tag = driver.find_elements(By.XPATH,'//div[@class="reviews"]')
for i in total_review_tag:
    j = i.text
    total_review.append(j)
over_tag = driver.find_elements(By.XPATH,'//div[@class="keyword"]')
for i in over_tag[2:32]:
    j = i.text
    over.append(j)
price_tag = driver.find_elements(By.XPATH,'//div[@class="price-col"]')
for i in price_tag:
    j = i.text
    price.append(j)
for i in price:
    if "Privates" in i:
        pp.append(i)
    else:
        dp.append(i)
    
view_tag = driver.find_elements(By.XPATH,'//a[@class="view-button"]')
for i in view_tag:
    j = i.get_attribute("href")
    urls.append(j)
    
driver.close()
for i in urls:
    driver = webdriver.Chrome(r"chromedriver.exe")
    driver.get(i)
    
    propd_tag = driver.find_elements(By.XPATH,'//div[@class="content collapse-content"]')
    for i in propd_tag[0:1]:
        j = i.text
        propd.append(j)
        
    facilities_tag = driver.find_elements(By.XPATH,'//ul[@class="facilities"]')
    for i in facilities_tag[0:1]:
        j = i.text
        fac.append(j)
        driver.close()
    
    
data = pd.DataFrame({"NMAE":name,"Distance":dist,
                    "Rating":rating,"Total Reviews":total_review,
                    "Drom Price":dp,"Private Price":pp,
                    "Facilities":fac,"Description":prpod})
data.to_csv("prop_london.csv")
 
```
