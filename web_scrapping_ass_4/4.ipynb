```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings("ignore")
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
driver = webdriver.Chrome(r"chromedriver.exe")
```

```python
driver.maximize_window()
```

```python
driver.get(r'https://github.com/')
```

```python
aa = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button')
aa.click()
```

```python
bb = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/ul[3]/li[3]')
bb.click()
```

```python
title = []
desc = []
count = []
language = []
asd = []
```

```python
try:
    title_tag = driver.find_elements(By.XPATH,'//h1[@class="h3 lh-condensed"]')
    for i in title_tag[0:23]:
        j = i.text
        title.append(j)
except NoSuchElementException as e:
    title.append("-")
```

```python
try:
    desc_tag = driver.find_elements(By.XPATH,'//p[@class="col-9 color-fg-muted my-1 pr-4"]')
    for i in desc_tag[0:23]:
        j = i.text
        desc.append(j)
except NoSuchElementException as e:
    desc.append("-")
```

```python
try:
    asd_tag = driver.find_elements(By.XPATH,'//a[@class="Link--muted d-inline-block mr-3"]')
    for i in asd_tag[0:46]:
        j = i.text
        asd.append(j)
except NoSuchElementException as e:
    asd.append("-")
```

```python
for i in asd:
    if asd.index(i) %2 !=0:
        count.append(i)
```

```python
try:
    lang_tag = driver.find_elements(By.XPATH,'//span[@class="d-inline-block ml-0 mr-3"]')
    for i in lang_tag[0:23]:
        j = i.text
        language.append(j)
except NoSuchElementException as e:
    language.append("-")
```

```python
data = pd.DataFrame({"TITLE":title,"DESCRIPTION":desc,
                    "COUNT":count,"LANGAUGE":language})
```
