```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings("ignore")
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome(r"chromedriver.exe")
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
driver.maximize_window()
```

```python
rank = []
name = []
artist = []
upload_date = []
views = []
```

```python
from selenium.common.exceptions import NoSuchElementException
```

```python
try:
    rank_tag = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
    for i in rank_tag[0:30]:
        j = i.text
        rank.append(j)
except NoSuchElementException as e:
    rank.append("-")
```

```python
try:
    name_tag = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')
    for i in name_tag[0:30]:
        j = i.text
        name.append(j)
except NoSuchElementException as e:
    name.append("-")
```

```python
try:
    artist_tag = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')
    for i in artist_tag[0:30]:
        j = i.text
        artist.append(j)
except NoSuchElementException as e:
    artist.append("-")
```

```python
try:
    upload_date_tag = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
    for i in upload_date_tag[0:30]:
        j = i.text
        upload_date.append(j)
except NoSuchElementException as e:
    upload_date.append("-")
```

```python
try:
    views_tag = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
    for i in views_tag[0:30]:
        j = i.text
        views.append(j)
except NoSuchElementException as e:
    views.append("-")
```

```python
len(name),len(rank),len(artist),len(upload_date),len(views)
```

```python
data = pd.DataFrame({"RANK":rank,"NAME":name,
                    "ARTIST":artist,"UPLOAD DATE":upload_date,
                    "VIEWS":views})
```
